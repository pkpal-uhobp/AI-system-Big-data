### 2.1. –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –º–∞—Ç–µ—Ä–∏–∞–ª ‚Äì –£—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä

**–°–∏–Ω—Ç–∞–∫—Å–∏—Å:**

```python

```

> üîπ –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:  
> –ö–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –¥–æ–ª–∂–µ–Ω –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ —É—Å–ª–æ–≤–∏—è, –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Å **–æ—Ç—Å—Ç—É–ø–æ–º –≤ 4 –ø—Ä–æ–±–µ–ª–∞** –æ—Ç —É—Ä–æ–≤–Ω—è `if`, `elif` –∏ `else`.  
> 
> –í Python –æ–±–ª–∞—Å—Ç–∏ –≤–∏–¥–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –æ–±–æ–∑–Ω–∞—á–∞—é—Ç—Å—è **–æ—Ç—Å—Ç—É–ø–∞–º–∏**.

### 2.3. –ó–∞–¥–∞–Ω–∏–µ

#### –ó–∞–¥–∞—á–∞:
–ù–∞–ø–∏—à–∏—Ç–µ –∫–æ–¥. –ó–∞–¥–∞—ë—Ç—Å—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è `x`. –ù–∞–ø–µ—á–∞—Ç–∞–π—Ç–µ, –∫ –∫–∞–∫–æ–º—É –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –∑–Ω–∞—á–µ–Ω–∏–µ `x`:

### 4.3.2. –ó–∞–¥–∞–Ω–∏–µ

#### –ó–∞–¥–∞—á–∞:
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è:  
$$
f(x) = \frac{\sqrt{1 + e^{\sqrt{x}} + \cos^2 x}}{|1 - \sin^3 x|} + \ln|2x|
$$

–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏($x$ –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ $ x $ –æ—Ç 1 –¥–æ 10).

–í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å:
- –ª–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –≤—Å–µ–≥–æ –º–∞—Å—Å–∏–≤–∞
- —Ç–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã (—Å—Ä–µ–∑–∞)

#### –†–µ—à–µ–Ω–∏–µ:

````python
import math
import numpy as np
import matplotlib.pyplot as plt
f = [(math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x ** 2))) / abs(1 - (math.sin(x)) ** 3) for x in range(1, 11)]
x = np.arange(0.0, max(f), max(f)/10)
newf = f[:5]
plt.grid()
plt.plot(x, f)
plt.show()
plt.grid()
plt.scatter(x[:5], newf)
plt.show()
     
````

### 4.3.3. –ó–∞–¥–∞–Ω–∏–µ

#### –ó–∞–¥–∞—á–∞:
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è:  
$$
f(x) = \left| \cos\left( x e^{\cos(x) + \ln(x+1)} \right) \right|
$$

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $(0; 10)$ —Å —à–∞–≥–æ–º $1$, –∑–∞–∫—Ä–∞—Å–∏—Ç—å –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –∫—Ä–∏–≤–æ–π –∏ –≤—ã—á–∏—Å–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–π –ø–ª–æ—â–∞–¥–∏. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(),
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

**–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ
–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:**
``` python
from scipy.integrate import simps
from numpy import trapz
```
#### –†–µ—à–µ–Ω–∏–µ:

````python
import math
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapezoid
f = [(abs(math.cos(x * (math.e ** (math.cos(x) + math.log(x + 1)))))) for x in range(0, 11)]
x = np.arange(0.0, max(f), max(f)/11)
plt.grid()
plt.plot(x, f, c = "r")
plt.fill_between(x, f)
area = trapezoid(f, x)
print(area)
````

`+`, `-`, `*`, `/`, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π:

````python
import math

print("–í–≤–µ–¥–∏—Ç–µ x: \n")
x = float(input())
print("–í–≤–µ–¥–∏—Ç–µ y: \n")
y = float(input())
print("–í–≤–µ–¥–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ:\n",
      "'+' (x + y)\n",
      "'-' (x - y)\n",
      "'*' (x * y)\n",
      "'/' (x / y)\n",
      "'e' (e ^ (x + y))\n",
      "'sin' (sin(x + y))\n",
      "'cos' (cos(x + y))\n",
      "'^' (x ^ y)\n",)
d = str(input())
if d == "+":
  print("x + y = ", x + y)
elif d == "-":
  print("x - y = ", x - y)
elif d == "*":
  print("x * y = ", x * y)
elif d == "/":
  if y == 0:
    print("–ù–∞ –Ω–æ–ª—å –¥–µ–ª–∏—Ç—å –Ω–µ–ª—å–∑—è!")
  else:
    print("x / y = ", x / y)
elif d == "e":
  print("e ^ (x + y) = ", math.e ** (x + y))
elif d == "sin":
  print("sin(x + y) = ", math.sin(x + y))
elif d == "cos":
  print("cos(x + y) = ", math.cos(x + y))
elif d == "^":
  print("x ^ y = ", x**y)
else:
  print("–û—à–∏–±–∫–∞!!!")
````
